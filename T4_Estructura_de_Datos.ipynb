Los conjuntos son otra estructura de datos similar a las listas y las tuplas pero más restrictivos. En este caso, los elementos deben ser inmutables, no se almacenarán elementos repetidos y no son accesibles por índice.

```python
# s4 : conjuntos
# se crean con { }

set0= set()
set0= {0}
set1= {1,2, "a"} # datos hetero
set1[0] # no permite acceder
```

```python
# s4 : conjuntos

set3= set ([1,1,2,3,4]) # no permite repetidios , solo muestra 1
set3
set4= {1,2,3,4,5}
set4.add(6)
# set4.add([1,2,3]) # no acepta incluir tipos avanzados
set4
set5= {1,2,3,4,5}
set5.update("E") # para incluir tipo cadena usamos update
set5
set5.remove("E")  # eliminar registro, si elemento no existe dara error por lo que podemos usar tb discard
set5
```

```python
# s4 : conjuntos

my_set = {"C++", "Python", "Java", "SQL"}

my_set.add("C") # Añadir un elemento
print(f"Add: {my_set}")
my_set.update(["Javascript", "React"]) # Añadir varios elementos
print(f"Update: {my_set}")
my_set.remove("C++")  # Elimina un elemento y genera un error si el elemento no existe
print(f"Remove: {my_set}")
my_set.discard("C++") # Elimina un elemento pero no genera error si el elemento no existe.
print(f"Discard: {my_set}")
my_set.clear()      # Elimina todos los elementos de un conjunto
print(f"Clear: {my_set}")
```

```python
# s4 : operaciones conjuntos NUEVA DIAPOSITIVA

a = {2,3}
b = {2,3,4}
a.intersection(b) #mismos valores en ambos
a.union(b) # unión de 2 conjuntos sin repetidos
a.symmetric_difference(b) # difernecia entre ambos
a.issubset(b) # si los elementos de a estan cotenidos en b, es un subconjunto de b
a.issuperset(b) # si todo b esta en a
```

Un diccionario es una estructura de datos en la que cada posición está representada por un par clave-valor. Se declaran con {}, al igual que con los conjuntos.

```python
# s4 : diccionarios
# se crean con { }

dict0 = dict()
dict0 = {}
dic1 = {"a" :1, "b" :2, "c": 3}
dict2 = dict ([("a",1), ("b",2), ("c",3)])
dict3= dict (a=1,b=2,c=3)
print(dic1["a"]) # elementos accesibles por claves , NO por indices
len (dic1) # len de las claves
print ("a" in dic1) # indica si está o no.
```

```python
# s4 : diccionarios

dic1 = {"a" :1, "b" :2, "c": 3}
# print (dic1[0:2]) no es una lista, al ser dict no es posible.
dic2 = dict(list(dic1.items())[:2]) # dic1.items Devuelve una vista de pares clave, lo transforma en lista, aplica slicing y convierte en dict
print(dic2)

#print(dic1 + {"d":4}) no puedes usar el operador + para combinar diccionarios
dic1["d"]=4
dic1.update({"e": 5})
print(dic1)


```

```python
# s4 : diccionarios

dic1 = {"a" :1, "b" :2, "c": 3}
dic2 = {"a" :4, "d" :5, "e": 6}

dic1.update(dic2)
print(dic1)

import operator

sorted(dic1.items(), key=operator.itemgetter(1)) # ordenar por valor
sorted(dic1.items(), key=operator.itemgetter(0)) # ordenar por clave

sorted(dic1.items(), key=operator.itemgetter(1), reverse = True) # ordenar por valor de mayor a menor
sorted(dic1.items(), key=operator.itemgetter(0), reverse = True) # ordenar por clave
```

```python
muestra = {
            "Cliente_1": {"Nacido": "Marzo", "Edad": 45},
            "Cliente_2": {"Nacido": "Mayo", "Edad": 15}
        }

print("Muestra:", muestra)

for value in muestra.values():
    if value["Nacido"] == "Marzo":
        value["Edad"] += 1

print("\nDict modificado:", muestra)

```

```python
def may_min(cadena: str):
    may = 0
    min = 0
    for char in cadena:
        if char.isupper():
            may += 1
        else:
            min += 1
    return may, min

print(may_min("Hola"))
```

```python
def func(*args: int):
    output = 0

    for i in args:
        if i % 2 == 0:
            i = i / 2
        output += i
                    
    return output
    

print(func(2,1, 4, 6))
```
