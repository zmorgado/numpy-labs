## Anlisis de correlaciones

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.concat([X, y], axis=1)
df.shape
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), cmap="coolwarm", center=0)
plt.title("Matriz de correlacin")
plt.show()
```

```python
df_malignant = df[df.target == 0]
df_malignant.shape
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df_malignant.corr(), cmap="coolwarm", center=0)
plt.title("Matriz de correlacin")
plt.show()
```

```python
df_benign = df[df.target == 1]
df_benign.shape
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df_benign.corr(), cmap="coolwarm", center=0)
plt.title("Matriz de correlacin")
plt.show()
```

```python
sns.boxplot(x="target", y="mean radius", data=df)
plt.title("Mean Radius por clase")
plt.show()
```

```python
sns.boxplot(x="target", y="mean area", data=df)
plt.title("Mean Area por clase")
plt.show()
```

```python
sns.histplot(data=df, x="worst concave points", hue="target", kde=True)
plt.title("Distribucin worst concave points por clase")
plt.show()
```

```python
sns.countplot(x="target", data=df)
plt.title("Distribucin de la variable objetivo")
plt.show()
```

```python
sns.violinplot(x="target", y="mean area", data=df)
plt.title("Mean Area por clase")
plt.show()
```
